```python
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split
import time
```

# Get CIFAR-10 datasets from torchvision.datasets and split

```python
transform = transforms.Compose([
    transforms.ToTensor(),
])

cifar_dataset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)

train_dataset, val_dataset = train_test_split(cifar_dataset, test_size=0.2, random_state=42)

test_dataset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
```

```python
len(train_dataset)
```

# Put them through a DataLoader

```python
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

```python

len(train_loader)
```

*italicized text*# Resnet18

```python

model_resnet = torchvision.models.resnet18(pretrained=True)
num_classes = 10
```

```python
model_resnet.fc = nn.Linear(model_resnet.fc.in_features, num_classes)

optimizer_resnet = torch.optim.SGD(model_resnet.parameters(), lr=0.001) #stochastic gradient descent
```

```python
criterion = nn.CrossEntropyLoss()
```

```python
num_epochs = 10
device = torch.device('cuda:0')
model_resnet.to(device)

start_time = time.time()
for epoch in range(num_epochs):
    model_resnet.train()
    for i, (images, labels) in enumerate(train_loader):
        images, labels = images.to(device), labels.to(device)

        optimizer_resnet.zero_grad()
        outputs = model_resnet(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer_resnet.step()

        if (i + 1) % 100 == 0:
            print(f'Epoch [{epoch + 1}/{num_epochs}], Step [{i + 1}/{len(train_loader)}], Loss: {loss.item():.4f}')

end_time = time.time()
training_time = end_time - start_time
print(f'Computation time: {training_time:.2f} seconds')
```

```python
model_resnet.eval()
correct = 0
total = 0
with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model_resnet(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f'Test Accuracy of the model on the test images: {accuracy:.2f} %')
```

# Densenet121

```python
model_densenet = torchvision.models.densenet121(pretrained=True)
num_classes = 10
```

```python
model_densenet.classifier = nn.Linear(model_densenet.classifier.in_features, num_classes)

optimizer_densenet = torch.optim.SGD(model_densenet.parameters(), lr=0.001)
```

```python
num_epochs = 10
device = torch.device('cuda:0')
model_densenet.to(device)

start_time = time.time()
for epoch in range(num_epochs):
    model_densenet.train()
    for i, (images, labels) in enumerate(train_loader):
        images, labels = images.to(device), labels.to(device)

        optimizer_densenet.zero_grad()
        outputs = model_densenet(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer_densenet.step()

        if (i + 1) % 100 == 0:
            print(f'Epoch [{epoch + 1}/{num_epochs}], Step [{i + 1}/{len(train_loader)}], Loss: {loss.item():.4f}')

end_time = time.time()
training_time = end_time - start_time
print(f'Computation time: {training_time:.2f} seconds')
```

```python
model_densenet.eval()
correct = 0
total = 0
with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model_densenet(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f'Test Accuracy of the model on the test images: {accuracy:.2f} %')
```
